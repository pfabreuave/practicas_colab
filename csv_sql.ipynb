```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("turma25.csv")
#print(df)
conn = sqlite3.connect("turma25.db")
df.to_sql("turma25_dataset", conn, if_exists="replace")
conn.close()
dbname = 'turma25.db'
conn = sqlite3.connect(dbname)
conn
read_df=pd.read_sql("select * from turma25_dataset", conn).drop(columns=["index"])
read_df
```
