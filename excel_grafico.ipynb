Crie um arquivo CSV a partir de um Excel para gerar um grÃ¡fico

```python
import pandas as pd
from matplotlib import pyplot as plt
from datetime import date
```

```python
today = date.today()

colors = [
    "#FFBA08",
    "#370617"
]
```

```python
read_file = pd.read_excel (r'turma25.xlsx')
read_file.to_csv (r'turma25.csv', index = None, header=True)
```

```python
df = pd.read_csv("turma25.csv")

depurado = df[ df["Ctrl_Proj"] >= 0] 

depurado = depurado.loc[:, ["Nome", "Nota_Pond"]].sort_values(by="Nota_Pond")

depurado = depurado.set_index("Nome")
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
plt.style.use("seaborn")
plt.barh(depurado.index, depurado["Nota_Pond"], color=colors)
plt.vlines(depurado["Nota_Pond"].mean(), 0, len(depurado), color="red", linestyles="dashed", label="Mediana")
plt.title(f"Registro de desempenho Turma 25 Ano {today.year}", fontweight="bold")
plt.xlabel("Nota ponderada obtida", fontweight="bold")
plt.ylabel("Nome do aluno", fontweight="bold")
plt.legend()
plt.tight_layout()
plt.show()
```
