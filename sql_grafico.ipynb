geração de gráficos a partir do banco de dados sqlite

```python
import sqlite3 as sql
import pandas as pd
from matplotlib import pyplot as plt
```

```python
colors = [
    
    "#FFBA08",

    "#370617"
]
```

```python
conn = sql.connect("tt25.db")
df = pd.read_sql('SELECT nome, nota_pond FROM t25_notas order by nota_pond', conn)
conn.commit()
conn.close()
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
df = df.set_index("nome")
plt.style.use("seaborn")
plt.barh(df.index, df["nota_pond"], color=colors)
plt.vlines(df["nota_pond"].mean(), 0, len(df), color="red", linestyles="dashed", label="Mediana")
plt.title("Registro de desempenho Turma 25", fontweight="bold")
plt.xlabel("Nota ponderada obtida", fontweight="bold")
plt.ylabel("Nome do aluno", fontweight="bold")
plt.legend()
plt.tight_layout()
plt.show()
```
